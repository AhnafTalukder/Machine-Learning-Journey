```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm      
from matplotlib.widgets import Slider
```

<h3>Training Examples</h3>

```python
x_train = np.array([1.0, 2.0])   #(size in 1000 square feet)
y_train = np.array([300.0, 500.0])         
```

<h3>Visualize Raw Data</h3>

```python
slope_w = 200 #adjust values for w and b to match the data
b = 100
x_range = np.linspace(x_train[0], x_train[x_train.shape[0]-1], 400)
y_range = [(x*slope_w)+b for x in x_range]

fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
ax.scatter(x_train, y_train, s = np.multiply(x_train, 100),c=[200, 200])
ax.plot(x_range, y_range)
ax.set_xlabel('Square Feet (1000s)')
ax.set_ylabel('Price in dollars')

plt.show( )
```

<h3>Cost Function Equation Implemented </h3>

```python
def compute_cost(x, y, w, b): 
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost    
```

<h3>Visualize Cost Function for a Range of w </h3>

```python
w_range = np.array([0, 400])
def visualize_cost(x, y, w_range, b_fixed):
    w_values = np.linspace(w_range[0], w_range[1], 400)
    #Create an empty  array
    temp_costs = []
    #loop through values of w and calculte cost
    for w in w_values:
        #cost = compute_cost(w)
        cost = compute_cost(x, y, w, b_fixed)
        #append value of cost to array
        temp_costs.append(cost)
        
    cost_values = np.array(temp_costs)
    fig, ax = plt.subplots()
    ax.plot(w_values, cost_values)
    plt.show()
```

```python
visualize_cost(x_train, y_train, w_range, 100)
```
